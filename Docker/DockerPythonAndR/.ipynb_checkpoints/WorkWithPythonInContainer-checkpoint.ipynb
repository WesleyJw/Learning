## How to read a dataset from a postgres container 

````python
#libs
import pandas as pd
import psycopg2
````

### Connect to postgres container

You need to provide some parameters to connect with postgres. The parameters are: host, user, dataset name and the password. Connections on the local machinne occur by setting host as localhost, but the in container service you to know the host. Frequentlly the host address can are  IPAddres of the container or container name. to know the container IPAddress of your postgres dataset you have to execute the follow comand:

```bash
docker inspect containerID | grep IPAddres
```

With all settings you can execute the connection on you database. 

````python
##Config variables

settings = {
    'host': '172.20.0.2',
    'user': 'postgres',
    'dbname': 'dvdrentail',
    'password': 'admin'
}
````

````python
conn = psycopg2.connect(**settings)
````

````python
data = pd.read_sql("SELECT * FROM category;", con = conn)
data
````

## Thing about: How many movies the actors works by the categories?  

To know the categories fimls we need the table: film_category, film_actor and film.

````python
pd.read_sql("SELECT * FROM film LIMIT 5;", con = conn)
````

````python
pd.read_sql("SELECT * FROM film_category LIMIT 5;", con = conn)
````

````python
pd.read_sql("SELECT * FROM film_actor LIMIT 5", con = conn)
````

````python
pd.read_sql("SELECT * FROM actor LIMIT 5", con = conn)
````

````python
data = pd.read_sql("""SELECT f.film_id, f.title, 
                        fc.film_id, fc.category_id,
                        ca.category_id, ca.name AS category,
                        fa.actor_id,
                        ac.first_name
                        FROM film f
                        LEFT JOIN film_category fc ON f.film_id = fc.film_id
                        LEFT JOIN category ca ON fc.category_id = ca.category_id 
                        LEFT JOIN film_actor fa ON  f.film_id = fa.film_id 
                        LEFT JOIN actor ac ON fa.actor_id = ac.actor_id """, con = conn)
data.head()
````

````python
data[['first_name', 'category', 'film_id']].groupby(['first_name', 'category']).count()
````
