## Análise Combinatória

### Introdução

O campo da matemática que é responsável por estudar os métodos e técnicas para resolver problemas de contagem é chamado de **análise combinatória** ou simplesmente **combinatória**. Sendo empregada principalmente em problemas que envolvem probabilidades, a combinatória faz análise das possibilidades e das combinações possíveis entre todos os elementos de um conjunto dado.

#### Princípio Fundamental da Contagem

Podemos definir o **princípio fundamental da contagem**, conhecido também como **princípio multiplicativo**, da seguinte forma:

*Quando um evento é composto por 'n' etapas sucessivas e independentes, de tal modo que as possibilidades da primeira etapa é 'z' e as possibilidades da segunda etapa é 't', temos que o número total de possibilidades do evento ocorrer é dado por meio do produto (z) . (t).*

Sendo assim, o princípio fundamental da contagem, resume-se em multiplicar o número de opções total entre as escolhas dadas.

Exemplo 1: 

André tem duas bermudas (Preta e Cinza) e quatro camisas (branca, verde, amarela e roxa). De quantas maneiras diferentes ele poderá se vestir usando uma bermuda e uma camiseta?

Veja na tabela as possibilidades (Diagrama de Árvore).

Bermudas | Camisas
---------|----------
P        | B
P        | v
P        | A
P        | R
C        | B
C        | V
C        | A
C        | R

Princípio multiplicativo: Quantidade de Opções de Bermudas x Quantidade de Opções de Camisas = 2 x 4 = 8

Exemplo 2:

Um restaurante oferece no cardápio 2 saladas distintas, 4 tipos de carne, 5 variedades de bebidas e 3 sobremesas diferentes. Uma pessoa deseja comer uma salada, uma carne, uma bebida e uma sobremesa. De quantas maneiras ela pode fazer seu pedido?

Tipo de Salada (S): 2
Tipo de Carne (C): 4
Variedade de Bebida (B): 5
Tipo de Sobremesa (SB): 3

Logo, temos que S x C x B x SB = 2 x 4 x 5 x 3 = 120 

#### Tipos de Combinatória

Com o princípio fundamental da contagem é possível resolver praticamente todos os problemas de contagem. Porém, seu uso se torna extremamente trabalhoso para alguns problemas do mundo real. .

Sendo assim, algumas técnicas foram desenvolvidas para ajudar a resolver estes problemas complexos. De certa forma a combinátoria pode ser classificada de três formas, sendo elas: arranjos, combinações e permutações.
